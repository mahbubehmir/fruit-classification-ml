```python
feature_name=['mass','width','height','color_score']
X=fruits[feature_name]
y=fruits['fruit_label']
cmap=cm.get_cmap('gnuplot')
Scatter=scatter_matrix(X, c=y,marker='o',s=40, hist_kwds={'bins':15}, figsize=(9,9),cmap=cmap)
plt.suptitle("scatterp_matrix for each input variable")
plt.savefig("fruits_scatter_matrix")
```

```python
X=fruits[["mass","width","height","color_score"]]
y=fruits['fruit_label']
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test, y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
#نرمالیزه کردن اطلاعات بین صفر و یک
from sklearn.preprocessing import MinMaxScaler
Scalar=MinMaxScaler()
X_train=Scalar.fit_transform(X_train)
X_test=Scalar.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

Logreg=LogisticRegression()
Logreg.fit(X_train,y_train)
print('Accuracy of Logistic regression classifier on training set:{:.2f}'.format(Logreg.score(X_train,y_train)))
print('Accuracy of Logistic regression classifier on test set:{:.2f}'.format(Logreg.score(X_test,y_test)))
Logreg.predict(X_train)
print(y_train)
Logreg.score(X_train,y_train)#میزان دقت
Logreg.predict(X_test)
y_test
Logreg.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
CIF=DecisionTreeClassifier().fit(X_train,y_train)

print('Accuracy of DecisionTressClassifier on training set:{:.2f}'.format(CIF.score(X_train,y_train)))
print('Accuracy of DecisionTressClassifier on test set:{:.2f}'.format(CIF.score(X_test,y_test)))
```

```python
dt.predict(X_train)
print(y_train)
dt.score(X_train,y_train)#میزان دقت
dt.predict(X_test)
y_test
dt.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
print('Accuracy of KNeighborsClassifier on training set:{:.2f}'.format(knn.score(X_train,y_train)))
print('Accuracy of KNeighborsClassifier on test set:{:.2f}'.format(knn.score(X_test,y_test)))
knn.predict(X_train)
print(y_train)
knn.score(X_train,y_train)#میزان دقت
knn.predict(X_test)
y_test
knn.score(X_test,y_test)
from sklearn.svm import SVC
svc=SVC()
```
